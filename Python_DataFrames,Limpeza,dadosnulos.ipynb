```python
# Traduzindo as categorias da coluna 'contrato'
#Prompt: Traduza para português as categorias da coluna contrato do dataframe DF

traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(traducao_contrato)

# Verificando o resultado
print("Contagem de valores após a tradução:")
print(df['contrato'].value_counts())
```

```python
# Traduzindo as categorias da coluna 'senioridade'
# Traduza para português as categorias da coluna senioridade do dataframe DF
traducao_senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(traducao_senioridade)

# Verificando o resultado
print("Contagem de valores da coluna 'senioridade' após a tradução:")
print(df['senioridade'].value_counts())
```

```python
# Traduzindo as categorias da coluna 'remoto'
#Traduza para português as categorias da coluna remoto do dataframe DF

traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(traducao_remoto)

# Verificando o resultado
print("Contagem de valores da coluna 'remoto' após a tradução:")
print(df['remoto'].value_counts())
```

```python
import pandas as pd
import numpy as np

#criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Fernanda", "Rafael", "Rita"],
    'salario': [10000, 20000, np.nan, np.nan]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
df_salarios
```

```python
import pandas as pd
import numpy as np

#criando um dataframe
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Fernanda", "Rafael", "Rita"],
    'salarios': [10000, 20000, np.nan, np.nan]
})

df_salarios["salario_media"] = df_salarios['salarios'].fillna(df_salarios["salarios"].mean().round(2))
df_salarios
```

```python
import numpy as np

#criando um dataframe
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Fernanda", "Rafael", "Rita"],
    'salarios': [10000, 20000, np.nan, np.nan]
})

#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios["salario_media"] = df_salarios['salarios'].fillna(df_salarios["salarios"].median())
df_salarios
```

```python
import numpy as np

#criando um dataframe
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Fernanda", "Rafael", "Rita", "Val"],
    'salarios': [10000, 20000, np.nan, np.nan, 100000]
})

#calcula a mediana salarial e substitui os nulos pela mediana
df_salarios["salario_media"] = df_salarios['salarios'].fillna(df_salarios["salarios"].median())
df_salarios
```

```python
#Pensar em um exemplo de monitoramento de temperatura
df_temperaturas = pd.DataFrame({
                "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
                "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
#ffill forward fill = completar com o valor anterior
```

```python
#Pensar em um exemplo de monitoramento de temperatura
df_temperaturas = pd.DataFrame({
                "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
                "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
#ffill forward fill = completar com o valor seguinte
```
